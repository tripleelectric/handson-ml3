```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# reproducible results
np.random.seed(42)
```

```python
def create_normal_data(rows: int, columns: int):
    '''
    1.  Normal data  (mean 0, var 1)
    '''
    return np.random.randn(rows, columns)
```

```python
X = create_normal_data(200, 2)
print(X.mean())
```

```python
# 3 sub‑plots: one for each of the three columns
fig, axs = plt.subplots(2, 1, figsize=(6, 8))

for i in range(2):
    axs[i].hist(X[:, i], bins=25, alpha=0.7, color='steelblue')
    axs[i].set_title(f'Column {i+1} – Histogram')
    axs[i].set_xlabel('Value')
    axs[i].set_ylabel('Frequency')

plt.tight_layout()
```

```python
# One scatter that shows all 3 columns against the row index
plt.figure(figsize=(8, 5))
plt.scatter(range(X.shape[0]), X[:, 0], color='tomato', label='col‑1', s=30)
plt.scatter(range(X.shape[0]), X[:, 1], color='royalblue', label='col‑2', s=30)

plt.title('Scatter of the 2 normal columns')
plt.xlabel('Row index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
